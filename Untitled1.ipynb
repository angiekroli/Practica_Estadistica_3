```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import math
```

```python
archivo_muertes = '/content/drive/MyDrive/Colab Notebooks/mat5.xlsx'
archivo_clima= '/content/drive/MyDrive/Colab Notebooks/Clima_Anual.xlsx'

df_muertes = pd.read_excel(archivo_muertes)
df_clima= pd.read_excel(archivo_clima)
#df_muertes=df_muertes.groupby(by="anio", dropna=False).sum()
df_muertes=df_muertes.merge(df_clima, how='inner', on='anio')
df_muertes=df_muertes.rename(columns={"muertos_en_accidentes_de_trafico_en_los_ultimos_10_anÌƒos": "muertes"})
df_muertes
```

```python
df_muertes.describe()
```

### **DENSIDAD DISTRIBUCIÃ“N PARA MUERTES 2013-2022**

```python
muertes=df_muertes["muertes"]
anio=df_muertes["anio"]
min_len,max_len= min(muertes),max(muertes)
n_bins=math.ceil((max_len-min_len)/20)
hist=np.histogram(muertes,bins=n_bins)
hist_dist=stats.rv_histogram(histogram=hist,density=True)
plt.hist(muertes,bins=n_bins,edgecolor="white",density=True)
eje_x=np.linspace(min_len,max_len,n_bins)
plt.plot(eje_x,hist_dist.pdf(eje_x),label="Linea Densidad")
X = stats.norm(92,18)
x = np.arange(X.ppf(0.01),X.ppf(0.99))
chi2 = stats.chi2(92)
gamma = stats.gamma(92)


plt.plot(x,X.pdf(x),"r",label="Dist.Normal")
plt.plot(x, chi2.pdf(x), "black",label="Dist Chi2")
plt.plot(x, gamma.pdf(x), "g",label="Dist Gamma")


plt.legend()
plt.title("Frecuencia muertes 2013-2022")
plt.xlabel("Muertes")
plt.ylabel("Frecuencia")
plt.show()



#La probabilidad de que haya 60 muertes en un mes es cerca de 3%, la probabilidad de que haya 120 muertes en un mes es del 1%
# La probabilidad tiene forma de distribuciÃ³n normal para el caso de muertes en accidentes
```

### **DENSIDAD DISTRIBUCIÃ“N PARA LLUVIA 2020,2021,2022**

```python
#Para ver la distribuciÃ³n del clima, voy a coger solo 3 aÃ±os (por comodidad y rapidez). Tengo dataframe de clima mensual. Solo cojo lluvia para el ejemplo.
archivo_clima_mes= '/content/drive/MyDrive/Colab Notebooks/Clima_mensual.xlsx'
df_clima_mes= pd.read_excel(archivo_clima_mes)
df_muertes=df_muertes[["mes_anio","muertes"]]
df_lluvia=df_clima_mes.merge(df_muertes, how='inner', on='mes_anio')
df_lluvia.describe()
```

```python
dias_lluvia=df_lluvia["Num_dias_lluvia"]
min_len,max_len= min(dias_lluvia),max(dias_lluvia)
n_bins=math.ceil((max_len-min_len)/2.5)
hist=np.histogram(dias_lluvia,bins=n_bins)
hist_dist=stats.rv_histogram(histogram=hist,density=True)
plt.hist(dias_lluvia,bins=n_bins,edgecolor="white",density=True)
eje_x=np.linspace(min_len,max_len,n_bins)
plt.plot(eje_x,hist_dist.pdf(eje_x),label="Linea Densidad")
X = stats.norm(8.85,3.54)
x = np.arange(X.ppf(0.01),X.ppf(0.99))
chi2 = stats.chi2(8.85)
gamma = stats.gamma(8.85)
#lognormal = stats.lognorm(8.85)

plt.plot(x,X.pdf(x),"r",label="Dist.Normal")
plt.plot(x, chi2.pdf(x), "black",label="Dist Chi2")
plt.plot(x, gamma.pdf(x), "g",label="Dist Gamma")
#plt.plot(x, lognormal.pdf(x), "yellow",label="Dist LogNormal")



plt.legend()
plt.title("Frecuencia lluvia 2020,2021,2022")
plt.xlabel("Dias lluvia")
plt.ylabel("Densidad")
plt.show()
```
